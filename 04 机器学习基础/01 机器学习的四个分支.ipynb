机器学习的核心难题：`过拟合`.
之前的3类机器学习都是`监督学习`(supervised learning)的例子,目标是学习训练输入与训练目标之间的关系。

## 机器学习的4个分支

#### 监督学习
监督学习是最常见的机器学习类型。监督学习主要包括分类和回归，但还有其他变体：
1. 序列生成：比如给定一张图，预测描述图像的文字
2. 语法树预测：给定一个句子，预测其分解生成的语法树
3. 目标检测：在图像中给目标画一个边界
4. 图像分割：给图像上特定物体画一个香酥鸡遮罩（mask）

#### 无监督学习
无监督学习是在没有目标的情况下寻找输入数据的变化。`降维`和`聚类`都是总所周知的无监督学习方法。

#### 自监督学习
自监督学习是监督学习的一个特例。自监督学习是没有人类参与的监督学习。标签仍然存在，但他们是从输入数据中生成。

#### 强化学习
在强化学习中，智能体（agent）接受环境信息，学会选择某种奖励最大话的行动。目前，强化学习主要集中在游戏上。未来的自动架势汽车，机器人等都会是强化学习的方向。

## 评估机器学习模型

评估模型的重点是将数据分为3个集合：训练集、验证集和测试集。如果可用数据少，有一下三种经典的评估方法：简单留出验证，K折验证，带有打乱数据的重复K折验证。

## 数据预处理、特征工程和特征学习
数据预处理的目的是使原始数据更适合于用神经网络处理，包括向量化，标准化、处理缺失数据和特征提取。

- 向量化
    - 素有输入和目标都必须是浮点数张量（特定情况下可以是整数张量）
- 值标准化
    - 取值较小，大部分值在0-1范围内
    - 同质性，所有特征值都在大致相同的范围内
    
## 过拟合与欠拟合
过拟合最优解决方法是获取更多训练数据。无法获取更多数据时，次有解决方法是调节模型允许存储的信息量，或对模型允许存储的信息加以约束。这种降低过拟合的方法叫做`正则化`(regularization)。
- 减小网络大小
    - 深度学习模型通常很擅长你和训练数据，但真正的挑战在于泛化，而不是拟合

## 机器学习通用工作流程

- 定义问题，收集数据集
    - 有一类非平稳问题，比如服装推荐引擎，在8月训练的数据在12月可能会有问题
- 选择衡量成功的指标
    - 平衡分类问题，精度和接受者操作特征去线下面积 是常用指标
    - 类别不平衡问题，可以使用准确率和召回率
    - 对于排序问题或多标签分类，可以使用平均准确率均值
    - 还可以自定义指标
- 确定评估方法
    - 留出验证集：数据量很大时采用
    - K折交叉验证：样本量太少，无法保证可靠性
    - 重复的k折验证：样本很少，同事模型评估又需要非常准确
- 准备数据
    - 数据格式化为张量
    - 张量取值通常缩放为较小的值
    - 如果不同的特征具有不同取值范围（异质数据），应该做数据标准化
    - 可能需要做特征工程，尤其是对于小数据问题
- 开发比基准更好的模型
    选择3个关键参数构建第一个工作模型：
    1. 最后一层的激活
    2. 损失函数，比如imdb的binary_crosssentropy，回归的例子使用mse等
    3. 优化配置，大多数情况下rmsprop及其默认的学习率是稳妥的
    ![img](images/004-001.png)
- 扩大模型规模：开发过拟合的模型
    理想的模型应该刚好在欠拟合和过拟合的节线上，在容量不足和容量过大的节线上
- 模型正则化与调节超参数
    这一步是最费时间的。需要不断调节模型、训练、在验证数据上评估，再次调节模型，然后重复过程
