### 概率建模

`概率建模`(Probabilistic modeling)是统计学在数据分析中的应用。它是最早的机器学习形式之一。最著名的算法就是朴素贝叶斯算法。

朴素贝叶斯算法(Naive Bayes)，假设输入数据的特征都是独立的。这是一个很强的假设，所以称之为`朴素`

logistic回归（logreg），被称为现代机器学习的“hello world”,注意不是一个回归算法，而是一个分类算法。

### 早期神经网络(Early neural networks)
贝尔实验室于1989年第一次实现了神经网络的实践应用，`LeNet`用于数字分类。90年代被美国邮政署采用，用于自动读取信封上的邮政编码

### 核方法 （Kernel Methods）

SVM: Support vector machine， 支持向量机。他的目标是在不同类别的2组数据点之间找到良好的决策边界（decision boundary）。
![%E5%9B%BE%E7%89%87.png](images/001-2-001.png)
SVM通过2步来寻找决策边界:
1. 将数据映射到一个新的高维表示
2. 间隔最大化（maximizing the margin）尽量数据点之间的距离最大化
![%E5%9B%BE%E7%89%87-3.png](images/001-2-002.png)

SVM缺点： 
- 很难扩展到大型数据集
- 在感知类问题上效果不好，比如图像分类等
    - 需要手动提取有用的特征工程，这一步骤很难，而且不稳定

### 神经网络
2012年以后，深度卷积神经网络已成为视觉任务的首选算法。更可以说，它在所有感知任务上都有效。

### 深度学习有何不同
发展迅速原因：
1. 在很多问题上表现出更好的性能
2. 解决问题更简单。它将工程完全自动化，这是机器学习最关键的一步

### 现状
- 梯度提升机，用于浅层学习问题
- 深度学习，用于感知问题
